```python
#the following analysis seeks to analyse which loan type is dominant in which region

region_loan_typelist=list(zip(regionlist,loan_typelist))

north_loantype_counts = {'type1': 0, 'type2': 0, 'type3': 0}
South_loantype_counts = {'type1': 0, 'type2': 0, 'type3': 0}
Central_loantype_counts = {'type1': 0, 'type2': 0, 'type3': 0}
northeast_loantype_count={'type1': 0, 'type2': 0, 'type3': 0}

for region, loantype in region_loan_typelist:
    if region == "North" and loantype in north_loantype_counts:
        north_loantype_counts[loantype] += 1
        
for region, loantype in region_loan_typelist:
    if region == "south" and loantype in South_loantype_counts:
        South_loantype_counts[loantype] += 1
        
for region, loantype in region_loan_typelist:
    if region == "central" and loantype in Central_loantype_counts:
        Central_loantype_counts[loantype] += 1
        
for region, loantype in region_loan_typelist:
    if region == "North-East" and loantype in northeast_loantype_count:
        northeast_loantype_count[loantype] += 1
        
    
print(north_loantype_counts)
print(South_loantype_counts)
print(Central_loantype_counts)
print(northeast_loantype_count)
```
